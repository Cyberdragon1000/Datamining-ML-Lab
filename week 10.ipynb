```python
################################################################################
#
#
# Your code to develop a decision tree classifier.
# Check the accuracy, performance metrices, and confusion metrics.
# Draw the decision tree using Graphviz.
# You can split your code into different cells according to your own choice
#
#
################################################################################

import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from IPython.display import Image 
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO
import pydotplus
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
%matplotlib inline
```

```python
df1 = pd.read_csv('car_evaluation\car_evaluation_0.csv')
df2 = pd.read_csv('car_evaluation\car_evaluation_1.csv')
df3 = pd.read_csv('car_evaluation\car_evaluation_2.csv')
df4 = pd.read_csv('car_evaluation\car_evaluation_3.csv')
# Your code to Print sample dataset
print(df1.shape)
df1.head()
```

```python
df=pd.concat([df1,df2,df3,df4])
dftest=pd.concat([df1,df2,df3,df4])
print(df.shape)
df.head()
```

```python
df.replace(to_replace=['low','med','high','vhigh','more','5more','big','small','unacc','acc','good','vgood'],value=[1,2,3,4,5,5,3,1,1,2,3,4],inplace=True)
df.head()
```

```python

labelencoder = LabelEncoder()
headers=['buyPrice','maintCost','noDoors','noPersons','bootLuggage','safety','decision']
for i in headers:
    labelencoder.fit(dftest[i])
    dftest[i] = labelencoder.transform(dftest[i])
    
dftest.head()

```

```python
# y = df.iloc[:,-1:]
# X = df.iloc[:,:-1]

X=df.drop('decision',axis=1).values
y=df['decision'].values

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = .10, random_state = 42)

print (X_train.shape)
print (X_test.shape)

print (y_train.shape)
print (y_test.shape)
print(y)
print(X)
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(X_train, y_train)
```

```python
feature_name = df.columns[:-1]
target_name = df.columns[-1]

print(feature_name)
print(target_name)
```

```python
dot_data = StringIO()  
tree.export_graphviz(clf, out_file= dot_data#'entropytree.dot'
                     ,feature_names=feature_name#,class_names=[target_name]
                     , filled=True, rounded=True, special_characters=True)  
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
```

```python
Image(graph.create_png())#,unconfined=True)
```

```python
y_pred1 = clf.predict(X_test)
print (classification_report(y_test, y_pred1))
cm1 = confusion_matrix(y_test, y_pred1)
disp1 = ConfusionMatrixDisplay(confusion_matrix = cm1, display_labels = clf.classes_)
disp1.plot()
plt.show()
```

```python
#loss
clf2 = DecisionTreeClassifier(criterion='log_loss')
clf2= clf2.fit(X_train, y_train)

dot_data2 = StringIO()  
tree.export_graphviz(clf2, out_file= dot_data2#'entropytree.dot'
                     ,feature_names=feature_name#,class_names=[target_name]
                     , filled=True, rounded=True, special_characters=True)  


graph2 = pydotplus.graph_from_dot_data(dot_data2.getvalue())  

Image(graph2.create_png())#,unconfined=True)
```

```python
y_pred2= clf2.predict(X_test)
print (classification_report(y_test, y_pred2))
cm2 = confusion_matrix(y_test, y_pred1)
disp2 = ConfusionMatrixDisplay(confusion_matrix = cm2, display_labels = clf2.classes_)
disp2.plot()
plt.show()
```

```python
#Gini
clf3 = DecisionTreeClassifier(criterion='gini')
clf3= clf3.fit(X_train, y_train)

dot_data3 = StringIO()  
tree.export_graphviz(clf3, out_file= dot_data3#'entropytree.dot'
                     ,feature_names=feature_name#,class_names=[target_name]
                     , filled=True, rounded=True, special_characters=True)  


graph3 = pydotplus.graph_from_dot_data(dot_data3.getvalue())  

Image(graph3.create_png())#,unconfined=True)
```
